```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_email(to_address, subject, body):
    from_address = 'your-email@example.com'
    password = 'your-email-password'

    msg = MIMEMultipart()
    msg['From'] = from_address
    msg['To'] = to_address
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'html'))

    try:
        server = smtplib.SMTP('smtp.example.com', 587)
        server.starttls()
        server.login(from_address, password)
        server.sendmail(from_address, to_address, msg.as_string())
        server.quit()
        print(f"Email sent to {to_address}")
    except Exception as e:
        print(f"Failed to send email: {e}")
```

```python
phishing_template = """
<html>
<head></head>
<body>
<p>Dear user,</p>
<p>We noticed unusual activity on your account. Please <a href="http://your-phishing-site.com?user={email}">click here</a> to verify your account.</p>
<p>Thank you,</p>
<p>Security Team</p>
</body>
</html>
"""

def create_phishing_email(email):
    return phishing_template.format(email=email)
```

```python
from flask import Flask, request, render_template_string

app = Flask(__name__)

@app.route('/phishing', methods=['GET', 'POST'])
def phishing():
    user = request.args.get('user')
    if request.method == 'POST':
        # Log the sensitive information submission
        sensitive_info = request.form.get('sensitive_info')
        log_submission(user, sensitive_info)
        return "Information submitted."
    # Log the click
    log_click(user)
    return render_template_string("""
    <form method="post">
        <label for="sensitive_info">Enter your sensitive information:</label>
        <input type="text" id="sensitive_info" name="sensitive_info">
        <input type="submit" value="Submit">
    </form>
    """)

def log_click(user):
    print(f"User {user} clicked the link.")

def log_submission(user, info):
    print(f"User {user} submitted: {info}")

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```

```python
def provide_education(user):
    print(f"Providing educational resources to {user}")

def log_click(user):
    print(f"User {user} clicked the link.")
    provide_education(user)

def log_submission(user, info):
    print(f"User {user} submitted: {info}")
    provide_education(user)
```

```python
from flask import Flask, render_template, request

app = Flask(__name__)

# In-memory storage for simplicity
users = []

@app.route('/')
def index():
    return render_template('index.html', users=users)

@app.route('/send_emails', methods=['POST'])
def send_emails():
    emails = request.form.get('emails').split(',')
    subject = "Security Alert"
    for email in emails:
        body = create_phishing_email(email.strip())
        send_email(email.strip(), subject, body)
        users.append({'email': email.strip(), 'clicked': False, 'submitted': False})
    return "Emails sent."

@app.route('/log_interaction', methods=['POST'])
def log_interaction():
    email = request.form.get('email')
    action = request.form.get('action')
    for user in users:
        if user['email'] == email:
            if action == 'click':
                user['clicked'] = True
            elif action == 'submit':
                user['submitted'] = True
    return "Interaction logged."

if __name__ == '__main__':
    app.run(debug=True)
```
