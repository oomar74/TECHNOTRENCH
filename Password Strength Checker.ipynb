```python
import re
import tkinter as tk
from tkinter import messagebox

def password_strength(password):
    # Define criteria for evaluation
    length_criteria = len(password) >= 8
    lower_criteria = re.search("[a-z]", password) is not None
    upper_criteria = re.search("[A-Z]", password) is not None
    digit_criteria = re.search("[0-9]", password) is not None
    special_criteria = re.search("[!@_#-$%^&*(),.?\":{}|<>]", password) is not None
    common_passwords = ["password", "123456", "123456789", "12345678", "12345", "1234567", "1234567890", "qwerty", "abc123", "letmein"]
    common_criteria = password.lower() not in common_passwords

    # Evaluate strength
    strength = 0
    if length_criteria: strength += 1
    if lower_criteria: strength += 1
    if upper_criteria: strength += 1
    if digit_criteria: strength += 1
    if special_criteria: strength += 1
    if common_criteria: strength += 1

    # Provide feedback
    if strength == 6:
        return "Strong", []
    elif 3 <= strength < 6:
        suggestions = []
        if not length_criteria: suggestions.append("Increase the length of your password to at least 8 characters.")
        if not lower_criteria: suggestions.append("Add lowercase letters to your password.")
        if not upper_criteria: suggestions.append("Add uppercase letters to your password.")
        if not digit_criteria: suggestions.append("Add numbers to your password.")
        if not special_criteria: suggestions.append("Add special characters to your password.")
        if not common_criteria: suggestions.append("Avoid using common passwords or patterns.")
        return "Moderate", suggestions
    else:
        return "Weak", ["Increase the length of your password to at least 8 characters.", "Add lowercase letters, uppercase letters, numbers, and special characters.", "Avoid using common passwords or patterns."]

def check_password():
    password = entry.get()
    strength, suggestions = password_strength(password)
    result_label.config(text=f"Password Strength: {strength}")
    if suggestions:
        messagebox.showinfo("Suggestions for Improvement", "\n".join(suggestions))

# Setup the GUI
root = tk.Tk()
root.title("Password Strength Checker")

tk.Label(root, text="Enter your password:").pack(pady=10)
entry = tk.Entry(root, show="*", width=30)
entry.pack(pady=10)

check_button = tk.Button(root, text="Check Password Strength", command=check_password)
check_button.pack(pady=10)

result_label = tk.Label(root, text="")
result_label.pack(pady=10)

root.mainloop()
```
